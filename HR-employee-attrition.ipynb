## Dataset attributes

1. **Age**: The employee's age in years.
2. **Attrition**: Indicates if the employee left the company ('Yes' or 'No'). **'Yes'**: they left the company, **'No'**: they didn't
3. **BusinessTravel**: Frequency of business travel for the employee (e.g., 'Travel_Rarely', 'Travel_Frequently').

4. **DailyRate**: The employee's daily pay rate. ($)

5. **Department**: The department the employee works in (e.g., 'Sales', 'Research & Development', 'Human Resources').

6. **DistanceFromHome**: The distance between the employee's home and work, typically in miles or kilometers. (*miles*)
- U.S. datasets typically use miles, while others might use kilometers).


7. **Education**: Employee's education level (e.g., 1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor').
8. **EducationField**: The field of study (e.g., 'Life Sciences', 'Medical', 'Marketing').

9. **EmployeeCount**: Constant value (1), often used as a placeholder. (*irrelevant*)
- In most cases, this column is a constant with a value of 1 for each employee, meaning it's a placeholder or irrelevant column. It doesn't provide any actionable data, and many analyses ignore it since it doesn't change across rows.


10. **EmployeeNumber**: A unique identifier for each employee.
11. **EnvironmentSatisfaction**: Satisfaction with the work environment (e.g., 1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High').
12. **Gender**: The employee's gender ('Male' or 'Female').

13. **HourlyRate**: The employee’s hourly wage. ($)
    
14. **JobInvolvement**: The employee's involvement in their job (e.g., 1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High').

This column likely categorizes employees by job hierarchy levels. The numbers probably correspond to organizational levels:

**1**: Entry-level or junior positions.

**2**: Mid-level positions.

**3**: Senior-level positions.

**4**: Executive or manager positions.

**5**: C-suite or top executive positions.

These levels typically reflect responsibility, pay grade, and seniority.



16. **JobLevel**: The employee's job level within the company.
17. **JobRole**: The employee's job role (e.g., 'Sales Executive', 'Research Scientist').
18. **JobSatisfaction**: Satisfaction with the job itself (e.g., 1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High').
19. **MaritalStatus**: The employee's marital status (e.g., 'Single', 'Married', 'Divorced').

20. **MonthlyIncome**: Refers to the actual amount the employee earns per month "the employee’s salary, *including* potential bonuses or incentives"
    

21. **MonthlyRate**: This column may represent the base rate or the standardized rate for the employee’s job role, possibly *before* bonuses or other incentives.

22. **NumCompaniesWorked**: The number of companies the employee has worked for in the past.
23. **Over18**: Whether the employee is over 18 years old ('Yes' or 'No').
24. **verTime**: Whether the employee works overtime ('Yes' or 'No').

25. **PercentSalaryHike**: This represents the percentage increase in salary over a defined period, likely over a *year* (as salary hikes are often calculated annually).
    
26. **PerformanceRating**: The employee’s performance rating (e.g., 1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding').
27. **RelationshipSatisfaction**: Satisfaction with workplace relationships (e.g., 1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High').

28. **StandardHours**: Standard working hours (usually constant value, often '40'). Constant Value of 80: This is ***unusual***
    

29. **StockOptionLevel**: The employee’s stock option level (e.g., 0: 'None', 1: 'Low', 2: 'Medium', 3: 'High').

This refers to the level of stock options the employee receives as part of their compensation package.

0: No stock options.

1: Low stock option grant.

2: Medium stock option grant.

3: High stock option grant.

Employees at higher job levels or with more seniority may receive more stock options as part of their compensation.


30. **TotalWorkingYears**: The total number of years the employee has worked in their career.
31. **TrainingTimesLastYear**: The number of training sessions the employee attended in the last year.
32. **WorkLifeBalance**: Employee’s work-life balance (e.g., 1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best').
33. **YearsAtCompany**: The number of years the employee has worked at the company.

34. **YearsInCurrentRole**: This refers to the number of years the employee has been in their current position or role within the company.
-  It helps understand employee tenure in a particular role, which could be valuable in analyzing attrition rates or job satisfaction over time.
   

35. **YearsSinceLastPromotion**: The number of years since the employee’s last promotion.
36. **YearsWithCurrManager**: The number of years the employee has worked with their current manager.

## Correlation

### Find the Correlation between the Numeric Features - using ***Pearsons Correlation Coefficients***

```python
df_nums = df.select_dtypes(include=['number'])

f, ax = plt.subplots(figsize=(30, 23))
sns.heatmap(df_nums.corr(), annot=True, linewidths=.5, fmt='.2f');
```

### Find the Correlation between the Target (Attrition) & Numeric Features - using ***Pearsons Correlation Coefficients***

```python
df_nums_corr = df_nums.corr()['Attrition'].abs().sort_values(ascending=False)

df_nums_corr
```

```python
top_features = df_nums_corr.head(17)

top_features
```

#### - We Choose the top 16 features most correlated with the target.

### Find the Correlation between the Target (Attrition) & Nominal Categorical Features - using ***Chi-Square***

```python
from scipy.stats import pointbiserialr
```

```python
corr, p_val = pointbiserialr(df['Attrition'], df['TrainingTimesLastYear'])
print("TrainingTimesLastYear")
print(f"P-value: {p_val:.6f}")
```

#### - We will use (0.022578) as the threshold, based on the least important numeric feature (TrainingTimesLastYear).

```python
from scipy.stats import chi2_contingency
```

```python
df_obj=df.select_dtypes(include='object').columns.tolist()

df_obj
```

```python
def chi_square_test(df, feature, target='Attrition'):
    contingency_table = pd.crosstab(df[feature], df[target])
    _, p_value, _, _ = chi2_contingency(contingency_table)
    return p_value

results = {}
for feature in df_obj:
    p_value = chi_square_test(df, feature)
    results[feature] = {'p-value': f"{p_value:.6f}"}

results_df = pd.DataFrame(results).T
results_df
```

#### - All categorical features have values below the threshold, so we will keep them all.
